```python
def convert_to_pdf(file_path, output_dir="output_files") -> Path:
    try:
        os.makedirs(output_dir, exist_ok=True)
        
        SOFFICE_PATH = r"C:\Program Files\LibreOffice\program\soffice.exe"

        subprocess.run([
            SOFFICE_PATH,
            "--headless",
            "--convert-to", "pdf",
            "--outdir", output_dir,  
            file_path
        ], check=True)

        original_filename = Path(file_path).stem 
        print(original_filename)
        new_filename = f"{original_filename}.pdf"
        
        return os.path.join(output_dir, new_filename)
    
    except Exception as e:
        print(f"Could Not Convert the Following File to .pdf: {e}")
        raise
```

```python
file_path = "pdfs/15. K Means clustering Algorithm.pptx"
output_dir="ouput_file"
```

```python
dir = convert_to_pdf(file_path,output_dir)
```

```python
async def save_files(file):
    upload_file= file
    os.makedirs(upload_file,exist_ok=True)
    file_path = upload_file/file.filename

    try:
        with open(file_path,"wb") as buffer:
            shutil.copyfileobj(file.file,buffer)
        print('Successfully saved file')
        return {"filepath":file_path,"filename":file.filename}
    
    except Exception as e:
        print("Unable to save file:",e)
        raise
```

```python
save_files(dir)
```
